# Over Sampling
oversampling involves randomly selecting examples from the minority class, with replacement, and adding them to the training dataset.

```python
df_test = pd.read_csv("Finance/test_data_hidden.csv")
df_test
```

```python
from numpy import mean
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
# define dataset
X, y = make_classification(n_samples=df_train.shape[0], weights=[0.99], flip_y=0)
# define pipeline
steps = [('over', RandomOverSampler()), ('model', DecisionTreeClassifier())]
pipeline = Pipeline(steps=steps)
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipeline, X, y, scoring='f1_micro', cv=cv, n_jobs=-1)
score = mean(scores)
print('F1 Score on Training dataset: %.3f' % score)
```

```python
# define dataset
X, y = make_classification(n_samples=df_test.shape[0], weights=[0.99], flip_y=0)
# define pipeline
steps = [('over', RandomOverSampler()), ('model', DecisionTreeClassifier())]
pipeline = Pipeline(steps=steps)
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipeline, X, y, scoring='f1_micro', cv=cv, n_jobs=-1)
score = mean(scores)
print('F1 Score on Testing Dataset: %.3f' % score)
```

# Undersampling

```python
# define dataset
X, y = make_classification(n_samples=df_train.shape[0], weights=[0.99], flip_y=0)
# define pipeline
steps = [('under', RandomUnderSampler()), ('model', DecisionTreeClassifier())]
pipeline = Pipeline(steps=steps)
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipeline, X, y, scoring='f1_micro', cv=cv, n_jobs=-1)
score = mean(scores)
print('F1 Score on Training Dataset: %.3f' % score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
dataset1 = pd.read_csv("Finance/train_data.csv")
dataset2 = pd.read_csv("Finance/test_data_hidden.csv")
```

```python
dataset1.shape
```

```python
dataset2.shape
```

```python
dataset = dataset1.append(dataset2)
dataset.shape
```

```python
corr_var=dataset.corr()
print(corr_var)
plt.figure(figsize=(20,30))
sns.heatmap(corr_var, annot=True, cmap='BuPu')
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=0)
print('Total no. of samples: Training and Testing dataset separately!')
print('X_train:', np.shape(X_train))
print('y_train:', np.shape(y_train))
print('X_test:', np.shape(X_test))
print('y_test:', np.shape(y_test))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier7 = LogisticRegression()
classifier7.fit(X_train,y_train)
```

```python
y_pred7 = classifier7.predict(X_test)
print(np.concatenate((y_pred7.reshape(len(y_pred7),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
cm7 = confusion_matrix(y_test, y_pred7)
print(cm7)
```

```python
from mlxtend.plotting import plot_confusion_matrix
fig, ax = plot_confusion_matrix(conf_mat=cm7, figsize=(6, 6), cmap=plt.cm.Greens)
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
logreg=accuracy_score(y_test,y_pred7)
logreg
```

```python
roc_auc_score(y_test, y_pred7)
```

```python
import sklearn.metrics as metrics
print(metrics.classification_report(y_test, y_pred7))
```

```python
import warnings
warnings.simplefilter("ignore")
from sklearn.model_selection import GridSearchCV
parameters_lr = [{'penalty':['l1','l2'],'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]
grid_search_lr = GridSearchCV(estimator = classifier7,
                           param_grid = parameters_lr,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search_lr.fit(X_train, y_train)
best_accuracy_lr = grid_search_lr.best_score_
best_paramaeter_lr = grid_search_lr.best_params_  
print("Best Accuracy of LR: {:.2f} %".format(best_accuracy_lr.mean()*100))
print("Best Parameter of LR:", best_paramaeter_lr)
```

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.20,random_state=1)
```

```python
trainY.shape
```

```python
import tensorflow as tf
from sklearn.preprocessing import Normalizer
transformer = Normalizer()
trainX = transformer.fit_transform(trainX)
testX = transformer.fit_transform(testX)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Reshape((trainX.shape[1],),input_shape=(trainX.shape[1],)))
```

```python
#Add 1st hidden layer
model.add(tf.keras.layers.Dense(2000))
model.add(tf.keras.layers.LeakyReLU())
model.add(tf.keras.layers.BatchNormalization())
```

```python
#Add 2nd hidden layer
model.add(tf.keras.layers.Dense(1000))
model.add(tf.keras.layers.LeakyReLU())
model.add(tf.keras.layers.BatchNormalization())
```

```python
#Add 3rd hidden layer
model.add(tf.keras.layers.Dense(1000))
model.add(tf.keras.layers.LeakyReLU())
model.add(tf.keras.layers.BatchNormalization())
```

```python
#Add 4th hidden layer
model.add(tf.keras.layers.Dense(300))
model.add(tf.keras.layers.LeakyReLU())
model.add(tf.keras.layers.BatchNormalization())
```

```python
#Add OUTPUT layer
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

# Learning rate (Adam)

```python
#Create optimizer with non-default learning rate
adam_op = tf.keras.optimizers.Adam(learning_rate=0.05)
#Compile the model
model.compile(optimizer=adam_op, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(trainX,trainY,          
          validation_data=(testX,testY),
          epochs=5,
          batch_size = 32)
```
